```python
# 11. Display descriptive statistics for the wage column, and the age column
df[['wage', 'age']].describe()
```

```python
# 12. Display descriptive statistics for the wage and age columns, by each race
df.groupby('race')[['wage', 'age']].describe()
```

```python
# 13. What do the mean and median values of wage tell us about the data?
print(df['wage'].mean())
print(df['wage'].median())
# The mean value of wage provides the average value of wage across the data set. 
# The median represents the middle value.
# The mean (111.70) is slightly higher than the median (104.92).
# This suggests that the distribution of wage is slightly skewed to the right.
# This means that some individuals have relatively higher wages or the potential presence of outliers. 
```

```python
# 14. Display the age and wage columns only
df[['wage', 'age']]
```

```python
# 15. Are these two columns correlated?
correlation_matrix = df[['wage', 'age']].corr()
correlation_coefficient = correlation_matrix.loc['wage', 'age']
correlation_coefficient
# THe correlation coefficient between the variables wage and age is close to 0 (0.19), which suggests a weak or no linear correlation.
```
